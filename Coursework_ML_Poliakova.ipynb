```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from matplotlib.ticker import AutoMinorLocator
import seaborn as sns
```

Определим несколько функций для вывода информации и построения графиков.

Функция для вывода информации по категориальным признакам:

```python
def print_counts(df, columns):
    
    """Функция распечатывает таблицу, содержащую value_counts количественно и в процентах, 
    для всех всех колонок из списка columns, входящих в датафрейм df"""
    
    for col in columns:
        
        counts = df[col].value_counts()
        percents = df[col].value_counts(normalize=True)
        result = pd.DataFrame({'count': counts, ' ': percents})
        
        # название колонки жирным шрифтом
        print(''.join(['\033[1m', col, '\033[0m']))
        # таблица
        print(result.to_string(float_format=lambda x: '{:.2%}'.format(x), justify='center'), end='\n\n')
```

Функция для аннотации горизонтальных столбчатых диаграмм (horizontal barchart):

```python
def annotate_barh(patches, text_array, format_spec, ax, ha='left', xytext=(2, 0)):
    """Функция выводит на график значения из списка text_array рядом со столбцами (bars)
    :patches: список объектов-столбцов
    :text_array: список значений, которые необходимо написать рядом со столбцами
    :ax: Axes, объект, который содержит график
    :format_spec, ha, xytext - параметры метода Axes.annotate()"""
    
    for bar, text in zip(patches, text_array):
        ax.annotate(format(text, format_spec),
                       (bar.get_width(), 
                        bar.get_y() + bar.get_height()/2), 
                      ha=ha, va='center', 
                      size=10, xytext=xytext, 
                      textcoords='offset points', alpha=0.8)
```

Функция для отрисовки расходящегося горизонтального барплота (diverging barplot) с аннотацией столбцов:

```python
def diverging_bar(data, ax=None, xlim=None):
    """Функция отрисовывает расходящийся барплот на основе датафрейма data размерности (n_rows, 2).
    По оси х отложено количество (counts), аннотации отражают процентное соотношение в каждом ряду.
    :data: DataFrame размерности (n_rows, 2)
    :ax: Axes, None по умолчанию. Объект, который будет содержать график. Если None, он будет создан автоматически
    :xlim: (x_min, x_max), None по умолчанию. Границы графика, если None, границы будут расчитаны автоматически"""
    
#     количество категорий
    ncat = data.shape[0]
#     данные в процентах (по рядам) для аннотации:
    data_percent = data.div(data.sum(axis=1), axis=0)

#     инициализация объекта Axes, на который будет наноситься график, если параметр ax=None
    if not ax:
        ax = plt.axes()

#     график
    data.mul([-1, 1]).plot.barh(stacked=True, ax=ax, alpha=0.9)
    
#     настройка оси х. нужно, чтобы аннотации не вылезали за рамки графика. границы задаются либо параметром xlim,
#     либо первоначальные границы расширяются на половину цены деления оси с обеих сторон
    if xlim:
        ax.set_xlim(xlim)
    else:
        xticks = ax.get_xticks()
        scale_interval = np.diff(xticks[:2])[0]
        x_min, x_max = ax.get_xlim()
        ax.set_xlim(x_min-scale_interval/2, x_max+scale_interval/2)
        
#     настройка отображения xticks. нужно, чтобы справа и слева от нуля стояли положительные числа
    ax.xaxis.set_major_formatter(lambda x, pos: np.abs(x, dtype='int64', casting='unsafe'))
#     добавление minor_ticks
    ax.xaxis.set_minor_locator(AutoMinorLocator(2))
    
#     аннотация столбцов слева от нуля
    annotate_barh(patches=ax.patches[:ncat], text_array=data_percent.iloc[:, 0].to_numpy(), format_spec='.0%', ax=ax, 
                  ha='right', xytext=(-2, 0))
#     аннотация столбцов справа от нуля
    annotate_barh(patches=ax.patches[ncat:], text_array=data_percent.iloc[:, 1].to_numpy(), format_spec='.0%', ax=ax, 
                  ha='left', xytext=(2, 0))
        
    ax.legend(title=data.columns.name, loc=2, bbox_to_anchor=(1, 1))
    ax.grid(ls='--', alpha=0.4, which='both')
```

Рассмотрим клиентов с подключенным интернетом. Подсчитаем количество и процент пользователей каждой услуги. Перед этим добавим в датасет колонку 'ZeroServices'. Она нужна для того, чтобы посчитать клиентов, у которых подключен интернет, но не подключено ни одной услуги.

```python
users_df.insert(14, 'ZeroServices', 
                ['Yes' if (row == 'No').all() 
                 else 'No' if (row == 'Yes').any() 
                 else 'No internet service' 
                 for i, row in users_df.iloc[:, 8:14].iterrows()])
```

```python
# список колонок, содержащих информацию об интернет-услугах
services = list(users_df.columns[8:15])

# выделим информацию о клиентах с подключенным интернетом в отдельный датафрейм
int_serv_df = users_df[users_df.InternetService != 'No'][['InternetService']+services]
```

```python
# преобразуем колонки так, что если клиент использует услугу, в соответствующей колонке будет стоять 1, если нет - 0
int_serv_df[services] = int_serv_df[services].apply(lambda x: x.factorize(sort=True)[0])
int_serv_df.head()
```

```python
# подсчитаем количество клиентов, пользующихся каждой из услуг, и отобразим на графике
servs_count = int_serv_df[services].sum().sort_values()

ax = servs_count.plot.barh(width=0.7)

annotate_barh(patches=ax.patches, text_array=servs_count.div(int_serv_df.shape[0]), format_spec='.0%', ax=ax)

ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.set_xlim((0, 2600))
ax.grid(axis='x', ls='--', alpha=0.5, which='both')
plt.show()
```

На графике выше показано, сколько клиентов, пользующихся интернетом через телекоммуникационную компанию, подключили ту или иную услугу. Так, например, видно, что примерно у 50% клиентов подключены услуги стримингового кино и телевидения, а у 13% не подключено ни одной услуги. Аналогичный график можно построить для каждой технологии подключения отдельно.

```python
# сгруппируем клиентов по технологиям подключения и подсчитаем, сколько людей пользуется каждой услугой
servs_by_tech = int_serv_df.groupby('InternetService')[services].sum().T.sort_values(by='Fiber optic')
# пересчитаем количество в проценты от общего количества клиентов DSL и Fiber optic
servs_by_tech_perc = servs_by_tech.div(int_serv_df.InternetService.value_counts())
servs_by_tech
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4), constrained_layout=True)

for ax, tech, c in zip(axes, ['Fiber optic', 'DSL'], ['tab:blue', 'tab:orange']):
    
    servs_by_tech[tech].plot.barh(width=0.7, ax=ax, color=c, alpha=0.9)
    
    annotate_barh(patches=ax.patches, text_array=servs_by_tech_perc[tech].to_numpy().T.ravel(), format_spec='.0%', ax=ax)
    ax.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax.set_title(tech)
    ax.grid(axis='x', ls='--', alpha=0.4, which='both')

axes[0].set_xlim((0, 1700))  
axes[1].set_xlim((0, 1150))
plt.show()
```

По графику выше видно, что профили подключенных услуг для двух технологий различаются. Количество пользователей DSL для каждой из услуг раличается не так сильно: наименее "популярная" услуга - StreamingTV - подключена примерно у 800 человек (40% клиентов DSL), а наиболее популярные - OnlineSecurity и TechSupport - подключены примерно у 1000 человек (48% клиентов). Разброс количества пользователей среди Fiber optic-клиентов значительно больше: OnlineSecurity подключена примерно у 700 человек (27%), а StreamingMovies - у 1500 человек (57%).

StreamingMovies и StreamingTV - самые популярные услуги среди клиентов Fiber optic и наименее популярные среди клиентов DSL. OnlineSecurity и TechSupport - наоборот, самые популярные среди клиентов DSL и наименее популярные среди клиентов Fiber optic.

Доля клиентов, у которых не подключено ни одной услуги (ZeroServices), приблизительно одинакова для обеих технологий и составляет 13% (Fiber optic) и 12% (DSL).

## 2.1. Категориальные признаки

```python
from scipy.stats.contingency import association
import itertools
```

Так как перед нами стоит задача предсказать отток клиентов, в первую очередь нас интересует, какие признаки непосредственно влияют на отток. Кроме того, интересно посмотреть на связь признаков между собой. Сложность задачи заключается в том, что целевая переменная и 16 из 19 признаков в нашем распоряжении - категориальные. Это значит, мы не можем построить pairplot и посмотреть на зависимости пар признаков. Так как практически все категориальные переменные в датасете относятся к номинальным переменным (nominal variables), мы можем воспользоваться мерой V Крамера, чтобы оценить степень связи признаков между собой. Значения V Крамера изменяются от 0 до 1, при этом 0 значит, что между переменными нет никакой связи, 1 обозначает идеальную ассоциацию. Результат можно нанести на тепловую карту.

```python
# список категориальных колонок и длина списка
categorical_cols = users_df.select_dtypes(['object']).columns.tolist()
n_cols = len(categorical_cols)

# массив, куда будем сохранять значения V Крамера
cramers_v_arr = np.empty(0)

# рассчитаем меру связи для каждой пары признаков
for col1, col2 in itertools.product(categorical_cols, repeat=2):
    
    # таблица сопряженности двух признаков
    contingency_table = pd.crosstab(users_df[col1], users_df[col2])
    # расчет V Крамера на основе таблицы сопряженности
    cramers_v = association(contingency_table, method='cramer')   
    # добавим полученное значение в массив cramers_v_arr
    cramers_v_arr = np.append(cramers_v_arr, cramers_v)

# преобразуем полученный массив в датафрейм размерности (n_cols, n_cols) для построения тепловой карты
heat_df = pd.DataFrame(cramers_v_arr.reshape(n_cols, n_cols), index=categorical_cols, columns=categorical_cols)
```

```python
# тепловая карта
plt.figure(figsize=(15, 15))
sns.heatmap(heat_df, annot=True)
plt.title("Cramer's V")
plt.show()
```

Есть несколько вещей, которые сразу бросаются в глаза:
* абсолюная ассоциация признаков PhoneService и MultipleLines. Действительно, зная значение MultipleLines, можно однозначно определить значение PhoneService. Это значит, что при построении алгоритма, точно можно будет отбросить признак PhoneService. Но и MultipleLines скорее всего нам не понадобится: оба признака практически не связаны с целевой переменной (значения коэффициентов < 0.05), а значит, в рамках задачи нет смысла изучать их зависимости.
    
    
* Практически нулевая ассоциация пола со всеми остальными признаками, включая целевую переменную. Это значит, что пол клиента не дает нам никаких инсайтов и ни на что не влияет. Нет смысла подробнее рассматривать зависимости, включающие пол, и уже на этом этапе можно сказать, что мы не будем ипользовать этот признак для построения алгоритма, предсказыающего отток.


* Достаточно сильная взаимная ассоциация (> 0.7) всех признаков, описывающих интернет-услуги. Эти зависимости стоит рассмотреть более детально.

Таким образом мы не позволим моделям воспринимать категориальные признаки как количественные и сравнивать категории между собой. ## 3.2. Подбор гиперпараметров моделей

[*Вернуться к содержанию*](#Содержание-работы)


Для подбора гиперпараметров моделей воспользуемся методом байессовской оптимизации, который реализован в библиотеке `hyperopt`. Лучшие значения гиперпараметров находятся постепенно. При первой итерации значения выбираются случайным образом из пространства параметров, которые определяет пользователь. Значения гиперпараметров в каждой следующей итерации выбираются с учётом результатов предыдущей так, чтобы минимизировать функцию потерь. Основная идея алгоритма заключается в поиске компромисса между исследованием регионов с самыми удачными из найденных комбинаций гиперпараметров и исследованием регионов с большой неопределённостью (где могут находиться ещё более удачные комбинации). Метод позволяет найти лучшие параметры быстрее, чем при переборе всех возможных комбинаций (Grid Search), и при этом вероятность упустить лучшую комбинацию значительно меньше, чем при случайном поиске (Random Search).

```python
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
```

### SVM

```python
# определим пространсво гиперпараметров
space4svm = {
    'C': hp.uniform('C', 0, 20),
    'kernel': hp.choice('kernel', ['linear', 'rbf']),
    'gamma': hp.uniform('gamma', 0, 20)
}

# определим целевую функцию, которую мы хотим минимизировать 
# она будет равна метрике качества с отрицательным знаком
# значение метрики качества на каждой итерации будет равно среднему, полученному на кросс-валидации
# функция будет возвращать отрицательный roc-auc score, статус итерации и модель

def f(params):
    clf = SVC(**params, class_weight='balanced')
    roc_auc = cross_val_score(clf, x_train, y_train, scoring='roc_auc', cv=5, n_jobs=-1).mean()
    return {'loss': -roc_auc, 'status': STATUS_OK, 'model': clf}    

# объект, в котором будут сохранены результаты всех итераций
trials_svm = Trials()

# поиск лучших гиперпараметров
best_hyperparams_svm = fmin(f, space4svm, algo=tpe.suggest, max_evals=100, trials=trials_svm)

# вывод результата на экран
print('best:')
print(best_hyperparams_svm)
```

### Logistic Regression

```python
space4lr = {
    'solver': hp.choice('solver', ['lbfgs', 'liblinear']),
    'C': hp.uniform('C', 0, 1)
}

def f(params):
    clf = LogisticRegression(**params, class_weight='balanced', random_state=37)
    roc_auc = cross_val_score(clf, x_train, y_train, scoring='roc_auc', cv=7, n_jobs=-1).mean()
    return {'loss': -roc_auc, 'status': STATUS_OK, 'model': clf}

trials_lr = Trials()

best_hyperparams_lr = fmin(f, space4lr, algo=tpe.suggest, max_evals=200, trials=trials_lr)

print('best:')
print(best_hyperparams_lr)
```

### Random Forest

```python
space4rf = {
    'max_depth': hp.choice('max_depth', range(1,15)),
    'max_features': hp.choice('max_features', range(1,10)),
    'n_estimators': hp.choice('n_estimators', range(1,200)),
    'criterion': hp.choice('criterion', ["gini", "entropy"]),
    'max_samples': hp.uniform('max_samples', 0, 1)
}

def f(params):
    clf = RandomForestClassifier(**params, class_weight='balanced', random_state=37)
    roc_auc = cross_val_score(clf, x_train, y_train, scoring='roc_auc', cv=7, n_jobs=-1).mean()
    return {'loss': -roc_auc, 'status': STATUS_OK, 'model': clf}

trials_rf = Trials()

best_hyperparams_rf = fmin(f, space4rf, algo=tpe.suggest, max_evals=200, trials=trials_rf)

print('best:')
print(best_hyperparams_rf)
```

### XGBClassifier

```python
# значение, которое соответствует сбалансированным весам классов
scale_pos_weight = (len(y)-sum(y)) / sum(y)

space4xgb={
    'max_depth': hp.choice("max_depth", range(1, 15)),
    'eta' : hp.uniform('eta', 0, 0.3),
    'gamma': hp.uniform('gamma', 1, 10),
    'reg_alpha' : hp.uniform('reg_alpha', 1, 180),
    'reg_lambda' : hp.uniform('reg_lambda', 0, 1),
    'subsample': hp.uniform('subsample', 0, 1),
    'colsample_bytree' : hp.uniform('colsample_bytree', 0, 1),
    'min_child_weight' : hp.choice('min_child_weight', range(0, 10)),
    'n_estimators': hp.choice('n_estimators', range(1, 200))
}

def f(params):
    clf = XGBClassifier(**params, 
                        objective='binary:logistic',
                        scale_pos_weight=scale_pos_weight, 
                        verbosity=0, use_label_encoder=False, 
                        seed=37)
    
    roc_auc = cross_val_score(clf, x_train, y_train, scoring='roc_auc', cv=7, n_jobs=-1).mean()
    return {'loss': -roc_auc, 'status': STATUS_OK, 'model': clf}

trials_xgb = Trials()

best_hyperparams_xgb = fmin(fn = f, space = space4xgb, algo = tpe.suggest, max_evals = 200, trials = trials_xgb)

print('best:')
print(best_hyperparams_xgb)
```

Best loss для моделей:

# 4. Сравнение качества моделей 

[*Вернуться к содержанию*](#Содержание-работы)

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_auc_score
```

Исходя из значений best loss, `XGBClassifier` показал наилуший результат. Протестируем модели на валидационной выборке. Получим их из объектов trials и обучим.

```python
# функция, получающая модель (не обученную) из объекта trials

def get_best_model_from_trials(trials):
    
    valid_trial_list = [trial for trial in trials
                            if STATUS_OK == trial['result']['status']]
    losses = [trial['result']['loss'] for trial in valid_trial_list]
    minumum_loss_index = np.argmin(losses)
    best_trial_obj = valid_trial_list[minumum_loss_index]
    
    return best_trial_obj['result']['model']
```

```python
# функция, выводящая roc_auc_score (с predict_proba) и матрицу ошибок

def print_model_results(model, x_test, y_test):
    
    y_pred = model.predict_proba(x_test)[:, 1]
    roc_auc = roc_auc_score(y_test, y_pred)
    
    print(f'\033[1m{type(model).__name__}\033[0m\nroc_auc = {roc_auc}\n')

    conf_mtrx = confusion_matrix(y_test, (y_pred > 0.5) * 1)
    ConfusionMatrixDisplay(conf_mtrx).plot(cmap='Blues')
    plt.show()
```

```python
fitted_models = []

for trials in [trials_svm, trials_lr, trials_rf, trials_xgb]:
    
    model = get_best_model_from_trials(trials)
    
    # для модели SVM необходимо поменять значение параметра probability с False на True
    # чтобы можно было вызвать метод predict_proba
    if type(model) == SVC:
        params = model.get_params()
        params.pop('probability')
        model = SVC(**params, probability=True)
        
    model.fit(x_train, y_train)
    fitted_models.append(model)
    
    print_model_results(model, x_valid, y_valid)
```

Судя по результатам, `Random Forest` и `XGBClassifier` справились с задачей лучше базовых алгоритмов. `XGBClassifier` показал лучший результат, чем `Random Forest`, и допустил меньше ошибок при определении минорного класса, поэтому сохраним `XGBClassifier` как лучшую модель:
