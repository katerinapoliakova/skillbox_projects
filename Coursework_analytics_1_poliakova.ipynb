* Минимальный, максимальный, средний, медианный возраст студентов:

Я буду рассчитывать возраст студентов на момент начала их обучения. Началом обучения будем считать дату начала выполнения первой домашней работы.

```python
# функция для отсеивания некорректных дат рождения

def is_birthday(x):
    if x is not np.nan and (1900 < int(str(x)[:4]) < 2019):
        return x
    else:
        return np.nan
```

```python
# отфильтруем и отсортируем основной датафрейм, оставим только первые дз студентов

study_start_dates = main[main.is_homework ==True].sort_values(by='start_date')\
.drop_duplicates(subset=['student_id', 'course_title'])[['course_title', 'student_id', 'start_date']]
study_start_dates.head()
```

```python
# отсортируем некорректные даты рождения в датасете students_df с помощью функции is_birthday, 
# удалим данные, с пустыми значениями дат,
# объединим полученный датасет с предыдущим (study_start_dates)
# таким образом получим датасет, содержащий название курса, id, дату рождения и дату начала обучения каждого студента;
# избавимся от лишних колонок
students_df['birthday'] = students_df.birthday.apply(is_birthday)
students_age_df = students_df.dropna(subset=['birthday'])\
.merge(study_start_dates, left_on='id', right_on='student_id')\
.drop(['id_', 'id', 'city'], axis=1)[['course_title', 'student_id', 'start_date', 'birthday']]

# добавим в датафрейм колонку age, которая будет содержать разницу даты начала обучения и даты рождения
students_age_df['age'] = (students_age_df.start_date.astype('datetime64[ns]')
                          - pd.to_datetime(students_age_df.birthday, format='%Y-%m-%d')).astype('timedelta64[Y]')

# избавимся от данных студентов с отрицательным значением возраста
students_age_df = students_age_df[students_age_df.age > 0]
students_age_df.head()
```

### Прогрессы по выполнению домашних работ

**При расчете прогрессов, будем учитывать незавершенные работы, находящиеся в статусе `start` или `fail`, как нагрузки только в первом месяце (когда началось выполнение работы).**

```python
# незавершенные работы учитываются, как нагрузки только в первом месяце

# отфильтруем основной датафрейм, оставив только данные, соответствующие выполнению домашних работ
# оставим колонки с названиями курсов, датами начала и конца выполнения домашних работ;
# избавимся от данных, для которых дата окончания дз наступила раньше даты начала выполнения дз
# приведем все даты к виду "год-месяц-01"

progress_in_months = main[main.is_homework == True][['course_title', 'start_date', 'finish_date']]
# progress_in_months = progress_in_months[(progress_in_months.start_date < progress_in_months.finish_date) | 
#                                         (np.isnat(progress_in_months.finish_date.astype('datetime64[ns]')))]
progress_in_months['start_date'] = progress_in_months.start_date.values.astype('datetime64[M]')
progress_in_months['finish_date'] = progress_in_months.finish_date.values.astype('datetime64[M]')

# для каждого месяца в указанный период добавим соответствующую колонку в датафрейм
# если студент выполнял домашнее задание в данном месяце, в соответствующей колонке будет стоять 1, если нет - 0

for date in pd.date_range('2016-03-01', '2019-07-01', freq='MS').values.astype('datetime64[M]'):
    progress_in_months[date] = [1 if (x <= date <= y) or 
                                (type(y) == type(pd.NaT) and x == date) 
                                 else 0 for x, y in zip(progress_in_months.start_date, progress_in_months.finish_date)]
progress_in_months
```

```python
# сгруппируем датафрейм по курсам, предварительно избавившись от колонок 'start_date', 'finish_date',
# и посчитаем количество прогрессов для каждого курса в каждом месяце

courses_progresses_df = progress_in_months.drop(['start_date', 'finish_date'], axis=1)\
.groupby('course_title')\
.sum()\
.stack()\
.reset_index()\
.rename(columns={'level_1': 'month', 0: 'progress_count'})
courses_progresses_df.tail()
```

### Минимальное, максимальное, среднее, медианное время прохождения модулей

Чтобы рассчитать время прохождения каждого модуля, нам нужно:

    1. определить, сколько домашних работ содержится в каждом модуле и найти среди них первую и последнюю домашние работы;
    2. получить данные о тех студентах, которые выполнили все домашние работы в модуле;
    3. найти для этих студентов дату начала выполнения первой и дату завершения последней домашней работы в модуле;
    4. вычислить длительность выполнения модулей.


```python
# Найдем количество домашних заданий в каждом модуле

homeworks_df = main[main.is_homework == True][['course_title', 'module_title', 'module_number', 'lesson_token']]
hw_count_df = homeworks_df.groupby(['course_title', 'module_title', 'module_number']).nunique().reset_index()\
.rename(columns={'lesson_token': 'hw_count'})
hw_count_df
```

```python
# определим первую и последнюю домашнюю работу в каждом модуле:
# оставим колонки, сожеджащие данные только о домашних работах, отсортируем датафрейм по номеру урока 

homeworks_df = main[main.is_homework == True][['course_title', 'module_number', 'lesson_number', 'lesson_token']]\
.drop_duplicates().sort_values(by='lesson_number')

# с помощью метода .drop_duplicates() оставим только первые дз в каждом модуле, а затем аналогично только последние, 
# сохранив токены соответствующих уроков в два отдельных датафрейма

first_hws_in_mod = homeworks_df.drop_duplicates(subset=['course_title', 'module_number'])\
.sort_values(by=['course_title', 'module_number'])[['course_title', 'module_number', 'lesson_token']]

last_hws_in_mod = homeworks_df.drop_duplicates(subset=['course_title', 'module_number'], keep='last')\
.sort_values(by=['course_title', 'module_number'])[['course_title', 'module_number', 'lesson_token']]
last_hws_in_mod.head()
```

```python
# посчитаем, сколько домашних работ выполнили студенты в каждом модуле, к которому приступали

done_hws = main[(main.is_homework == True) & (main.start_date < main.finish_date)]\
[['course_title', 'module_title', 'module_number', 'student_id', 'lesson_token']]\
.groupby(['student_id', 'course_title', 'module_title', 'module_number']).count().reset_index()\
.rename(columns={'lesson_token': 'hw_count'})
done_hws
```

```python
# оставим данные только о тех студентах, которые выполнили все домашние работы в модуле:
# объединим датафрейм, содержащий информацию о количестве дз в каждом модуле (hw_count_df), 
# с датафреймом, содержащим данные о количестве дз, выполненных студентами, в каждом модуле (done_hws);
# только тогда, когда количество дз в модуле (hw_count_df.hw_count) совпадает с количеством дз, 
# которое выполнил студент в данном модуле (done_hws.hw_count), можно считать, что студент выполнил все работы

full_mod_students = hw_count_df.merge(done_hws, on=['course_title', 'module_title', 'module_number', 'hw_count'])
full_mod_students
```

```python
# объединим полученный датафрейм с основным, чтобы получить даты и токены уроков для студентов, выполнивших все задания

full_mod_students = full_mod_students.merge(main[['course_title', 'module_title', 'module_number', 'student_id', 
                                                  'start_date', 'finish_date', 'lesson_token']],
                                            on=['course_title', 'module_title', 'module_number', 'student_id'])
full_mod_students
```

```python
# с помощью полученных датафреймов first_hws_in_mod и last_hws_in_mod найдем даты начала выполнения первого дз,
# и даты завершения последнего дз в модуле для нашей группы студентов;

filtered_for_dur_start = full_mod_students\
.merge(first_hws_in_mod, on=['course_title', 'module_number', 'lesson_token'], how='right')\
[['course_title', 'module_number', 'module_title', 'student_id', 'start_date']] # датафрейм с началом первых работ в модуле

filtered_for_dur_finish = full_mod_students\
.merge(last_hws_in_mod, on=['course_title', 'module_number', 'lesson_token'], how='right')\
[['course_title', 'module_number', 'module_title', 'student_id', 'finish_date']] # датафрейм с окончанием последних работ в модуле

# объединим два датафрейма в итоговый;
# start_date будет соответствовать дате начала выполнения первой работы в модуле,
# finish_date - дате завершения последней работы в модуле

module_duration_df = filtered_for_dur_start.merge(filtered_for_dur_finish, 
                                                  on=['course_title', 'module_number', 'module_title', 'student_id'])

# разницу между окончанием и началом прохождения модуля в секундах запишем в колонку duration

module_duration_df['duration'] = (module_duration_df.finish_date - module_duration_df.start_date).astype('timedelta64[s]')
module_duration_df
```

```python
# сгруппируем данные по курсу и номеру модуля и получим необходимые показатели с помощью метода .describe()

module_duration_stat = module_duration_df.groupby(['course_title', 'module_number'])['duration']\
.describe()[['min', 'max', 'mean', '50%']]\
.reset_index()\
.sort_values(by=['course_title', 'module_number'])\
.rename(columns={'50%': 'dur_median'})

module_duration_stat
```

```python
# тот же датафрейм, где время содержится в более наглядном формате

module_duration_stat_days = module_duration_stat.copy()
module_duration_stat_days[['min', 'max', 'mean', 'dur_median']] = module_duration_stat[['min', 'max', 'mean', 'dur_median']]\
.apply(lambda x: pd.to_timedelta(x, unit='s'))
module_duration_stat_days.sort_values(by='max')
```

### Сезонность

Чтобы оценить сезонность по длительности выполнения домашних работ, я буду относить домашнюю работу к тому месяцу, в котором началось ее выполнение, таким образом, я получу медианное время выполнения работ, начатых в данном месяце.

```python
# выделим данные по завершенным домашним работам, отфильтровав некорректные даты;
# сохраним длительность выполнения работ в отдельную колонку

dur_in_months = main[(main.is_homework == True) & (main.start_date < main.finish_date)]\
[['course_title', 'start_date', 'finish_date']]
dur_in_months['duration'] = (dur_in_months.finish_date - dur_in_months.start_date).astype('timedelta64[s]')
dur_in_months.head()
```

```python
# добавим в датафрейм колонки для каждого месяца в году, куда будем сохранять:
# длительность выполнения работы, если выполнение работы началось в данном месяце, или NaN в противном случае 

for i in range(1, 13):
    dur_in_months[str(i)] = [dur if start.month == i else np.nan for dur, start in zip(dur_in_months.duration,
                                                                                       dur_in_months.start_date)]
dur_in_months.head()
```

```python
# избавимся от лишних колонок в датафрейме, сгруппируем данные по названию курса;
# найдем медианное время выполнения домашней работы в каждом месяце
# если медианное время == NaN, заменим пустое значение на ноль

dur_in_months = dur_in_months.drop(['start_date', 'finish_date', 'duration'], axis=1).groupby('course_title')\
.median()\
.stack(dropna=False)\
.reset_index()\
.rename(columns={'level_1': 'month', 0: 'dur_median'})
dur_in_months
```

```python
import calendar

plt.figure(figsize=(15,10))
plt.title('Медианное время выполнения домашних работ в зависимости от месяца', fontsize=14)
plot = sns.lineplot(data=dur_in_months, x='month', y='dur_median', hue='course_title', linewidth=2)
plot.spines['top'].set_visible(False)
plot.spines['right'].set_visible(False)
yticks = pd.Series(range(0, int(dur_in_months.dur_median.max() + 24*3600), 24*3600))
plt.yticks(yticks, pd.to_timedelta(yticks, unit='s').dt.days)
plt.xticks(dur_in_months.month.unique(), [calendar.month_name[i] for i in range(1, 13)])
plt.xlabel(None)
plt.ylabel('Медианное время, в днях')
plt.legend(title='Курс', loc='upper left', bbox_to_anchor=(0.05, 0.97), framealpha=0.6)
plt.grid(alpha=0.5)
# plot.set_frame_on(False) ## вариант с полным отсутствием рамок
plt.show()
```

## 4. Расчет конверсии

1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).
2. **(m)** Постройте bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. График должен иметь заголовок. Ticks нужно развернуть так, чтобы они были читаемы
3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐
4. На основании рассчитанных значений сделайте аналитический вывод (должен быть полным и покрывать все полученные выше метрики)

```python
# чтобы расчитать конверсию перехода студентов из одного модуля в другой, по условию задания, нам нужны данные студентов,
# начавших первое задание в модуле и завершивших последнее;
# для этого воспользуемся полученными ранее датафреймами, которые содержат токены соответствующих уроков

filtered_start = main[(main.start_date < main.finish_date) | (main.finish_date.isna())]\
.merge(first_hws_in_mod, how='right')[['course_field', 'course_title', 'module_number', 'module_title', 'student_id',
                                       'start_date']]

filtered_finish = main[(main.start_date < main.finish_date) | (main.finish_date.isna())]\
.merge(last_hws_in_mod, how='right')[['course_field', 'course_title', 'module_number', 'module_title', 'student_id', 
                                      'finish_date']]

# объединим датафреймы в один

filtered_for_conv = filtered_start.merge(filtered_finish, 
                                         on=['course_field', 'course_title', 'module_number', 'module_title', 'student_id'])
filtered_for_conv.head()
```

```python
# сгруппировав данные по названию курса и номеру модуля, подсчитаем количество студентов, начавших (колонка started)
# и завершивших (колонка finished) каждый модуль; сохраним эти данные в новый датафрейм
    
module_conversion_df = filtered_for_conv\
.groupby(['course_field', 'course_title', 'module_number', 'module_title'])[['start_date', 'finish_date']]\
.count().reset_index().rename(columns={'start_date': 'started', 'finish_date': 'finished'})
module_conversion_df.head()
```

```python
# расчитаем конверсию для каждого курса, результат будем записывать в список conv_list

conv_list = []
for name, course in module_conversion_df.groupby('course_title', sort=False):
    conv_list.append(1.0)    # конверсия первого модуля в курсе равна 1, поэтому при смене курса в цикле, сохраняем 1.0 в список
    for i in course.index[1:]: # конверсия для остальных модулей в курсе
        if  course.loc[i-1].finished == 0: # если предыдущий модуль завершило 0 человек, конверсия стремится к бесконечности,
            conversion = np.nan            # сохраняем NaN
        else:
            conversion = course.loc[i].started / course.loc[i-1].finished
        conv_list.append(conversion)

# сохраним полученный список в колонку conversion

module_conversion_df['conversion'] = conv_list
module_conversion_df
```

```python
# Построим bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе

plt.figure(figsize=(18,7))
plt.title('Конверсия перехода студентов из одного модуля в другой на различных курсах', fontsize=14)
r = sns.barplot(data=module_conversion_df, x='course_title', y='conversion', hue='module_number', palette='viridis')
plt.tick_params(axis='x', labelrotation=90, labelsize=12)
plt.xlabel('')
plt.ylabel('Конверсия', fontsize=12)
plt.legend(title='Номер модуля', ncol=3, loc=2, bbox_to_anchor=(0.01, 0.98))
plt.show()
```

По графику видно, что для большинства модулей конверсия остается в пределах единицы. Заметное исключение составляют 11 модуль курса "UX-дизайн" и 6 модуль курса "SMM-маркетолог от А до Я". Более наглядное представление можно получить, если построить графики отдельно для каждого курса.

```python
# графики конверсии для каждого курса

field_colors = {f: c for f, c in zip(module_conversion_df.course_field.unique().tolist(), sns.color_palette('Set2')[:4])}
medians = module_conversion_df.groupby('course_title', sort=False)['conversion'].median()
height_ratios = module_conversion_df.groupby('course_title', sort=False)['module_number'].count().tolist()

fig, axes = plt.subplots(15, 1, figsize=(15,50), constrained_layout=True, gridspec_kw={'height_ratios': height_ratios})
fig.suptitle('Конверсия перехода студентов из одного модуля в другой на различных курсах', fontsize=16, x=0.5, y=1.017)

for course, ax in zip(module_conversion_df.course_title.unique(), axes):
    ax.set_title(course, fontsize=14, loc='left')
    data = module_conversion_df[module_conversion_df.course_title == course]
    sns.barplot(data=data, x='conversion', y='module_title', ax=ax, orient='h', 
                color=field_colors.get(data.iloc[0, 0]), alpha=0.9)
    
    ax.axvline(medians[course], ls='--')
    
    ax.set_yticks(range(0, data.shape[0]))
    # названия некоторых модулей очень длинные, поэтому, чтобы графики смотрелись лучше, укоротим названия до 50 символов
    ax.set_yticklabels(data[['module_title', 'module_number']]\
                       .agg((lambda x: x[0][:50] + '... ' + str(x[1]) if len(x[0]) > 50 else x[0] + ' ' + str(x[1])), axis=1))
    ax.set_ylabel(None)
    ax.set_xlabel(None)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
#     ax.set_frame_on(False) ## вариант с полным отсутствием рамок

handles = [axes[i].get_children()[0] for i in [0, 2, 7, 13]] + [axes[1].get_children()[32]]
labels = module_conversion_df.course_field.unique().tolist() + ['Median']
fig.legend(handles, labels, loc=9, ncol=5, bbox_to_anchor=(0.5, 1.012), fontsize=12, title_fontsize=12)
plt.show()
```

## 5. Метрика успеваемости 

&nbsp;&nbsp;&nbsp;&nbsp;*(необязательное задание)*

Иногда студенты берут курсы, которые оказываются для них неподъемными. Это может быть как по причинам недостаточной изначальной подготовки, так и по причинам, связанным с низкой мотивацией студента. Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

В качестве метрики успеваемости студента я выбрала соотношение количества выполненных домашних работ и длительности обучения. Если курс оказался слишком трудным для студента, то он либо не будет выполнять домашние работы, либо их выполнение будет занимать длительное время; то же касается студентов с низкой мотивацией. Поэтому проблемными студентами будут считаться те, кто учится на курсе уже достаточно долго, например, дольше полугода, но сдал меньше двух домашних работ.

```python
# крайняя дата в датафрейме
max_date = main[['start_date', 'finish_date']].max().max()
```

```python
# главный датафрейм, очищенный от некорректных данных о начале и завершении домашних работ
main_filtered = main[(main.is_homework == True) & ((main.start_date < main.finish_date) | (main.finish_date.isna() == True))]
```

```python
# нам нужны даты начала обучения каждого студента, 
# поэтому отсортируем датафрейм по start_date и оставим только первые строки для каждого прогресса

all_students_df = main_filtered[['course_title', 'progress_id', 'student_id', 'start_date']]\
.sort_values(by='start_date')\
.drop_duplicates(subset='progress_id')
```

```python
# подсчитаем количество завершенных домашних работ для каждого студента на каждом курсе
# добавим соответствующую колонку в датафрейм

prog_hw_count = main_filtered.groupby('progress_id')['finish_date']\
.count().reset_index().rename(columns={'finish_date': 'hw_count'})

all_students_df = all_students_df.merge(prog_hw_count, on='progress_id')
```

```python
# условие для выбора проблемных студентов: выполнено менее 2 работ при времени обучения более полугода (185 дней)

problem_students_df = all_students_df[(all_students_df.hw_count < 2) & 
                                      ((max_date - all_students_df.start_date) > pd.Timedelta('185 days'))]\
.drop(['start_date', 'hw_count'], axis=1)
problem_students_df
```

Выше представлен список проблемных студентов с учетом id прогресса и курса.

Мы также можем рассчитать долю проблемных студентов в целом на платформе и на каждом курсе в отдельности.

```python
# доля проблемных студентов на всех курсах

print(np.round((problem_students_df.progress_id.nunique() / main.progress_id.nunique())*100, 2), '%')
```

```python
# доля проблемных студентов на каждом курсе

# количество проблемных студентов на каждом курсе:
problem_students_s = problem_students_df.course_title.value_counts()\
.reset_index().rename(columns={'index': 'course_title', 'course_title': 'problem_sts'})

# доля проблемных студентов на каждом курсе:
problem_students_perc = student_count_df.merge(problem_students_s, on='course_title', how='left')
problem_students_perc['percent'] = np.round(problem_students_perc['problem_sts']/problem_students_perc['student_count'], 4) * 100
problem_students_perc.sort_values(by='percent', ascending=False)
```
